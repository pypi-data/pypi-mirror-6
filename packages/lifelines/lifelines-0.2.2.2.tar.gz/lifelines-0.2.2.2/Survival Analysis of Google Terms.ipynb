```python
import vertdb
import pandas as pd
import lifelines.estimation as estimation
```

```python
data = vertdb.query("""SELECT 
    DATEDIFF('dd', created_at, COALESCE( churned_at, CURRENT_DATE() ) ) as 'lifetime',   
    (churned_at IS NOT NULL)::INT as 'observed',
    term as 'key'
FROM shopify.shops s
JOIN funnel.new_leads l 
  on l.shop_id = s.shop_id
JOIN "multitrack"."conversion_visits" c 
  on c.shop_id = s.shop_id
JOIN multitrack.visit_facts vf
  on vf.visit_id = c.visit_id
  and vf.term IS NOT NULL
where DATE(created_at) >= '2013-02-01' AND
term IN
( 
     SELECT term
  FROM "multitrack"."conversion_visits" cv
 JOIN multitrack.visit_facts vf 
  on vf.visit_id = cv.visit_id
WHERE term IS NOT NULL and DATE(vf.timestamp) >= '2012-01-01' and category in ('adwords')
 and term NOT ILIKE '%shop%fy%'
 GROUP BY term
 ORDER BY COUNT(1) DESC
 LIMIT 30
)
""")
```
