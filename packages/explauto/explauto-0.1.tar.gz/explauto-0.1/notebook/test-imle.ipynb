```python
from skel_xp_torso import Torso

#torso = Torso(os='linux', opti_addr='193.50.110.196')
torso = Torso(os='linux', dummy_robot=True)
```

```python
#data = load('clement.npy')
#data = vstack([load('../data/torso-trajectoiries/record_{}.npy'.format(i)) for i in range(1, 5)])
data = load('../data/model.npy')
#data = load('../data/data_clean.npy')

#random.shuffle(data)

data_train = data[:15000,:]
data_test = data[15000:19000,:]

q_train = data_train[:, :7]
pos_train = data_train[:, 7:]

q_test = data_test[:, :7]
pos_test = data_test[:, 7:]

noise_value = 0.01
noise_train = noise_value * randn(len(pos_train), 3)
#noise_test = noise_value * randn(len(pos_test), 3)

pos_train += noise_train
#pos_test += noise_test
```

```python
e_knn=[]
regr = KNeighborsRegressor(n_neighbors=20)
regr.fit(pos_train, q_train)
for p in pos_test:
    q=regr.predict(p)
    p2 = translation_from_transf(torso.right_arm_chain.forward_kinematics(q)[0])
    e_knn.append(norm(p2 - p))

print mean(e_knn)
plot(e_knn)
```

```python
model = imle.Imle(alpha=0.995,Psi0=[noise_value**2] * 3, sigma0=pi/24, wsigma=2**7, wSigma=2**7, wNu=7,wLambda=0.1, wPsi=1000, p0=0.1, multiValuedSignificance=0.8, nSolMax=6)
#model = imle.Imle(Psi0=[noise_value**2] * 3,  p0=0.1, wsigma=0.1, wSigma=0.1, wNu=0.1, wLambda=0.1, wPsi=2**7, nSolMax=5, multiValuedSignificance=0.8)
#model = imle.Imle(p0=0.1, wsigma=0.1, wSigma=0.1, wNu=0.1, wLambda=0.1, wPsi=0.1, nSolMax=10, multiValuedSignificance=0.1)
```

```python
for i, (z, x) in enumerate(zip(q_train, pos_train)):
    if i % 1000 == 0:
        print i
    
    model.update(z, x)
```

```python
print model.get_joint_mu(4)
```

```python
from pypot.kinematics import translation_from_transf

e = []
nsol = []
sol = []
weight = []
qss=[]

for p in pos_test:
    qs = model.predict_inverse(p)
    q=qs[argmax(model.get_prediction_weight())]
    qss.append(qs)
    weight.append(model.get_prediction_weight())
    es = []
    nsol.append(len(qs))
    #for q in qs:
    p2 = translation_from_transf(torso.right_arm_chain.forward_kinematics(q)[0])
    es.append(norm(p2 - p))
        
    e.append(min(es))
    sol.append(argmin(es))
    #weight.append(argmax(model.get_prediction_weight()))
```

```python
inds=find(array(nsol)>1)
print weight[inds[0]]
print qss[inds[0]]
print pos_test[0]
print translation_from_transf(torso.right_arm_chain.forward_kinematics(qss[inds[0]][0])[0])
print translation_from_transf(torso.right_arm_chain.forward_kinematics(qss[inds[0]][1])[0])
```
